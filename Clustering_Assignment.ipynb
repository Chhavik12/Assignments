# Theory Questions

```python
#1 What is unsupervised learning in the context of machine learning.
#AnsUnsupervised learning is a type of machine learning that discovers hidden patterns, relationships,
#or structures within unlabeled data, where no predefined outputs or target variables are provided.
```

```python
#2 How does K-Means work?
#Ans
#Select the number of clusters (k).
#Randomly initialize the centroids for each cluster.
#Assign each data point to the nearest centroid.
#Recalculate each centroid as the mean of the points assigned to it.
#Repeat steps 3 and 4 until the centroids no longer change (convergence).
```

```python
#3 Explain the concept of a dendrogram in hierarchical clustering
#AnsA dendrogram is a tree-like diagram that illustrates how data points or clusters merge or split at 
#different distance levels during hierarchical clustering.
```

```python
#4 Difference between K-Means & Hierarchical Clustering
#Ans
#K-Means: Forms flat clusters, requires the number of clusters (k) to be specified beforehand, and is computationally faster.
#Hierarchical Clustering: Builds a tree-like hierarchy of clusters, does not require k initially, but is generally slower and more computationally intensive.
```

```python
#5 What are the advantages of DBSCAN over K-Means
#Ans
#No need to specify the number of clusters (k) in advance.
#Can detect clusters of arbitrary shapes.
#Robust to noise and outliers, handling them effectively.
```

```python
#6 When would you use Silhouette Score in clustering
#Ans
#It is used to measure the quality of clustering, evaluating how well each data point has been assigned to its cluster compared to others.
```

```python
#7 What are the limitations of Hierarchical Clustering
#Ans
#Computationally expensive with a complexity of O(n2)O(n^2)O(n2).
#Sensitive to noise and outliers, which can distort cluster formation.
#Difficult to correct incorrect merges or splits once they occur in the hierarchy.
```

```python
#8  Why is feature scaling important in clustering algorithms like K-Means
#Ans
#Because hierarchical clustering relies on distance metrics (like Euclidean distance),
#using unscaled features can distort the cluster formation, giving undue weight to features with larger ranges.
```

```python
#9 How does DBSCAN identify noise points
#Ans
#Points that do not belong to any dense region—having fewer than minPts within a radius ε—are classified as noise in DBSCAN.
```

```python
#10 Define inertia in the context of K-Means
#AnsThe sum of squared distances between each data point and its cluster centroid, which measures how compact the clusters are.
```

```python
#11  What is the elbow method in K-Means clustering
#Ans
#Plot inertia vs. k; the “elbow” point indicates optimal number of clusters (diminishing improvement after that).
```

```python
#12 Describe the concept of "density" in DBSCAN
#Ans
#Density refers to the number of points within a specified radius (ε). Regions with high density are considered clusters, while low-density areas are treated as noise.
```

```python
#13 Can hierarchical clustering be used on categorical data
#AnsYes, hierarchical clustering can handle categorical data if appropriate distance measures
#(e.g., Gower distance or Hamming distance) are used, but it is not suitable with standard Euclidean distance.
```

```python
#14 What does a negative Silhouette Score indicate
#Ans
#A point is likely misclassified if it is closer to a different cluster than the one it has been assigned to.
```

```python
#15 Explain the term "linkage criteria" in hierarchical clustering
#AnsIt defines the method used to compute distances between clusters in hierarchical clustering.
#Common linkage methods include: Single, Complete, Average, and Ward linkage.
```

```python
#16 Why might K-Means clustering perform poorly on data with varying cluster sizes or densities
#Ans Because K-Means assumes clusters are spherical, of similar size, and have similar density,
#it performs poorly on irregularly shaped or unevenly sized clusters.
```

```python
#17 What are the core parameters in DBSCAN, and how do they influence clustering
#Ans
#ε (Eps): The radius defining a point’s neighborhood.
#minPts: The minimum number of points required to form a dense region.
#Together, they determine cluster density and noise sensitivity in DBSCAN.
```

```python
#18 How does K-Means++ improve upon standard K-Means initialization
#Ans Chooses initial centroids smartly — spreads them apart to reduce poor clustering and speed up convergence
```

```python
#19 What is agglomerative clustering
#Ans It is a bottom-up hierarchical clustering method that starts with each data point as an individual cluster and iteratively merges clusters based on similarity.
```

```python
#20 What makes Silhouette Score a better metric than just inertia for model evaluation
#Ans The Silhouette score evaluates clustering by considering both cohesion (how close points are within the same cluster)
#and separation (how far points are from other clusters), providing a measure of overall cluster quality.
```

```python
#21 Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
#scatter plot
#Ans 
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='black', marker='X')
plt.title("K-Means Clustering (4 Centers)")
plt.show()
```

```python
#29  Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart
#Ans
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=400, centers=4, random_state=42)
scores = []

for k in range(2, 6):
    km = KMeans(n_clusters=k, random_state=41).fit(X)
    scores.append(silhouette_score(X, km.labels_))

plt.bar(range(2, 6), scores, color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()
```

```python
#30  Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage
#Ans 
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

iris = load_iris()
Z = linkage(iris.data, method='average')

plt.figure(figsize=(10, 5))
dendrogram(Z, labels=iris.target)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
#31 Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries
#Ans 
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

# Decision boundary
x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='rainbow')
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200)
plt.title("K-Means with Decision Boundaries (Overlapping Clusters)")
plt.show()
```

```python
#34   Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot
#Ans 
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = load_breast_cancer().data
inertias = []

for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=41).fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(2, 7), inertias, marker='o')
plt.title("Inertia vs K (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.show()
```

```python
#46 Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features)
#Ans 
# ==============================
# Wine Dataset - KMeans Clustering Visualization
# ==============================

import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# ✅ Fix Windows MKL warning (optional)
os.environ["OMP_NUM_THREADS"] = "1"

# 1. Load the Wine dataset
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# 2. Standardize features
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# 3. Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df_scaled['Cluster'] = kmeans.fit_predict(df_scaled)

# 4. Print size of each cluster
print("Cluster sizes:")
print(df_scaled['Cluster'].value_counts())

# 5. Visualize using Seaborn Pairplot (first 5 features for clarity)
sns.pairplot(
    df_scaled[df_scaled.columns[:5].tolist() + ['Cluster']],
    hue='Cluster',
    palette='Set2'
)
plt.suptitle("K-Means Clustering on Wine Dataset", y=1.02)
plt.show()
```
