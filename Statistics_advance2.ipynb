```python
#Question 1: What is hypothesis testing in statistics?
```

```python
#Ans 
#Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves:
#Formulating an assumption (hypothesis) about a population parameter.
#Collecting sample data.
#Using statistical tests to decide whether the evidence supports or rejects the assumption.
```

```python
#Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
```

```python
#Ans
#Null Hypothesis (H₀): A statement that assumes there is no effect, no difference, or status quo in the population.
#Example: “The mean height of students is 170 cm.”
#Alternative Hypothesis (H₁ or Ha): A statement that contradicts the null hypothesis, claiming there is an effect or difference.
#Example: “The mean height of students is not 170 cm.”
#Difference: H₀ represents the claim to be tested, while H₁ represents what we conclude if the evidence is strong enough to reject H₀.
```

```python
#Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
```

```python
#Ans
#Significance Level (α): The threshold probability of rejecting H₀ when it is actually true (Type I error).
#Common values: 0.05, 0.01.
#Role: If the p-value ≤ α, reject H₀; otherwise, fail to reject H₀.
```

```python
#Question 4: What are Type I and Type II errors? Give examples of each.
```

```python
#Ans
#Type I Error (False Positive): Rejecting H₀ when it is actually true.
#Example: Concluding a medicine works when it doesn’t.

#Type II Error (False Negative): Failing to reject H₀ when H₀ is false.
#Example: Concluding a medicine doesn’t work when it actually does.
```

```python
#Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
```

```python
#Ans
#| Feature                       | **Z-Test**                     | **T-Test**                     |
#| ----------------------------- | ------------------------------ | ------------------------------ |
#| Population Std Dev (σ) known? | Yes                            | No (use sample std dev)        |
#| Sample size                   | Large (n ≥ 30)                 | Small (n < 30)                 |
#| Distribution                  | Normal                         | t-distribution                 |
#| Usage example                 | Comparing mean of large sample | Comparing mean of small sample |
```

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
#Python. Show the Python code and interpret the results.
#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
#50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
#50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
#50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population parameters
mu_0 = 50   # Null hypothesis mean
sigma = 0.5 # Assumed population std deviation
alpha = 0.05

# Sample statistics
mean_sample = np.mean(sample_data)
n = len(sample_data)

# Z-statistic
z_stat = (mean_sample - mu_0) / (sigma / np.sqrt(n))

# Critical Z for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# P-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean: {mean_sample:.4f}")
print(f"Z-Statistic: {z_stat:.4f}")
print(f"Z-Critical (two-tailed): ±{z_critical:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision
if abs(z_stat) > z_critical:
    print("Reject H₀: Significant difference from population mean.")
else:
    print("Fail to reject H₀: No significant difference from population mean.")
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data
np.random.seed(42)  # For reproducibility
mean_true = 50
std_dev = 5
n = 100
data = np.random.normal(mean_true, std_dev, n)

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation

# 95% confidence interval
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

margin_of_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.4f}")
print(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")

# Plot
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"Lower 95% CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle='dashed', linewidth=2, label=f"Upper 95% CI = {ci_upper:.2f}")
plt.title("Normal Distribution Data with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores represent
#in terms of standard deviations from the mean.


import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean_val = np.mean(data)
    std_val = np.std(data)
    z_scores = (data - mean_val) / std_val
    return z_scores

# Generate dataset
np.random.seed(42)
data = np.random.normal(100, 15, 200)  # mean=100, std=15, n=200

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Print first 5 Z-scores
print("First 5 Z-scores:", z_scores[:5])




#Explanation of Z-scores:
#Z-scores tell us how many standard deviations a data point is from the mean.
#Z = 0 → value is exactly at the mean.
#Z = 1 → value is 1 standard deviation above the mean.
#Z = -2 → value is 2 standard deviations below the mean.
```
